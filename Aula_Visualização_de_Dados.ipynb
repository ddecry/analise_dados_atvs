Desafio 1.1: Comparação de Vendas por Região

Cenário: Uma empresa de varejo quer comparar o desempenho de vendas de suas quatro regiões no último trimestre para identificar qual delas precisa de mais atenção.



```python
# Dados para o Desafio 1.1
dados_regiao = {
    'Regiao': ['Norte', 'Sul', 'Leste', 'Oeste'],
    'Vendas_Milhoes': [12.5, 18.2, 15.0, 9.8]
}
df_regiao = pd.DataFrame(dados_regiao)
print("\nDataFrame para Desafio 1.1:")
print(df_regiao)
```

```python
# Escreva sua justificativa aqui:
# 1. Tipo de gráfico escolhido: Gráfico de Barras
#
# 2. Justificativa:
# Eu escolhi um gráfico de barras porque ele é um dos melhores para comparar valores entre categorias.
# Nesse caso, queremos comparar as vendas de cada região.
#
# Pelo modelo de Cleveland-McGill, comparar valores usando o comprimento das barras em um mesmo eixo é uma forma bem precisa de visualizar diferenças.
#
# Pela Gestalt, as barras são parecidas e ficam organizadas lado a lado, o que facilita enxergar qual região vendeu mais e qual vendeu menos.
# Além disso, seguindo a ideia do Tufte, o gráfico de barras é simples e direto, sem muitos elementos desnecessários, deixando a informação principal clara.

# Seu código Python para o gráfico:
plt.figure(figsize=(8, 5))
sns.barplot(data=df_regiao, x='Regiao', y='Vendas_Milhoes')

plt.title("Vendas por Região no Último Trimestre")
plt.xlabel("Região")
plt.ylabel("Vendas (em milhões)")

plt.show()
```

Desafio 1.2: Distribuição de Idades de Clientes

Cenário: Uma equipe de marketing quer entender a distribuição de idade de seus clientes para segmentar campanhas. Eles têm dados de idade de 1000 clientes.

```python
# Dados para o Desafio 1.2
np.random.seed(42)
idades = np.random.normal(loc=35, scale=10, size=1000).astype(int)
idades = idades[(idades >= 18) & (idades <= 70)] # Limitar idades razoáveis
df_idades = pd.DataFrame({'Idade': idades})
print("\nDataFrame para Desafio 1.2 (primeiras 5 linhas):")
print(df_idades.head())
```

```python
# Escreva sua justificativa aqui:
# 1. Tipo de gráfico escolhido: Histograma
#
# 2. Justificativa:
# Eu escolhi um histograma porque ele é o gráfico mais comum para entender a distribuição de uma variável
# numérica, como a idade. Ele ajuda a ver em quais faixas de idade existem mais clientes e se os dados
# estão mais concentrados em uma parte ou espalhados.
#
# Pelo modelo de Cleveland-McGill, o histograma funciona bem porque a gente compara o tamanho das barras
# e consegue perceber rapidamente quais faixas de idade aparecem mais.
#
# Pela Gestalt, as barras ficam agrupadas e organizadas em ordem crescente, então nosso cérebro entende
# facilmente que aquilo representa uma sequência (idades aumentando).
#
# E seguindo a ideia do Tufte, o histograma é simples e direto, mostrando só o necessário para entender
# o comportamento das idades, sem colocar informações extras que podem atrapalhar.

# Seu código Python para o gráfico:
plt.figure(figsize=(8, 5))
sns.histplot(df_idades['Idade'], bins=20)

plt.title("Distribuição de Idades dos Clientes")
plt.xlabel("Idade")
plt.ylabel("Quantidade de Clientes")

plt.show()
```

Desafio 1.3: Tendência de Acessos ao Site

Cenário: A equipe de produto quer monitorar a evolução diária do número de acessos ao site nos últimos 30 dias para identificar padrões ou quedas.

```python
# Dados para o Desafio 1.3
datas = pd.date_range(start='2024-03-01', periods=30)
np.random.seed(42)
acessos = np.random.randint(1000, 5000, 30)
# Simular uma queda nos últimos dias
acessos[-5:] = np.random.randint(500, 1500, 5)
df_acessos = pd.DataFrame({'Data': datas, 'Acessos': acessos})
print("\nDataFrame para Desafio 1.3 (primeiras 5 linhas):")
print(df_acessos.head())
```

```python
# Escreva sua justificativa aqui:
# 1. Tipo de gráfico escolhido: Gráfico de Linha
#
# 2. Justificativa:
# Eu escolhi um gráfico de linha porque ele é o mais indicado quando queremos analisar mudanças ao longo do tempo.
# Como os dados são acessos por dia, esse tipo de gráfico ajuda a enxergar facilmente se houve crescimento,
# queda ou algum padrão durante o mês.
#
# Pelo modelo de Cleveland-McGill, a comparação pela posição dos pontos no eixo vertical (altura da linha)
# é bem precisa, então conseguimos perceber rapidamente quando os acessos aumentam ou diminuem.
#
# Pela Gestalt, a linha conecta os pontos e cria a sensação de continuidade, o que faz nosso cérebro entender
# que os valores fazem parte de uma sequência temporal.
#
# E seguindo a ideia do Tufte, o gráfico de linha é simples e direto, sem enfeites, mostrando claramente
# a tendência dos acessos e destacando a queda nos últimos dias.

# Seu código Python para o gráfico:
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_acessos, x='Data', y='Acessos')

plt.title("Acessos Diários no Site (Últimos 30 dias)")
plt.xlabel("Data")
plt.ylabel("Número de Acessos")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Desafio 1.4: Relação entre Investimento em Marketing e Vendas

Cenário: A equipe de finanças quer entender se há uma correlação entre o investimento mensal em marketing e o volume de vendas gerado no mesmo mês.

```python
# Dados para o Desafio 1.4
np.random.seed(42)
investimento_marketing = np.random.randint(10000, 50000, 12)
vendas_mensais = investimento_marketing * 5 + np.random.randint(-5000, 5000, 12)
df_marketing_vendas = pd.DataFrame({
    'Investimento_Marketing': investimento_marketing,
    'Vendas_Mensais': vendas_mensais
})
print("\nDataFrame para Desafio 1.4 (primeiras 5 linhas):")
print(df_marketing_vendas.head())
```

```python
# Escreva sua justificativa aqui:
# 1. Tipo de gráfico escolhido: Gráfico de Dispersão (Scatter Plot)
#
# 2. Justificativa:
# Eu escolhi um gráfico de dispersão porque ele é o mais indicado quando queremos ver a relação
# entre duas variáveis numéricas. Nesse caso, queremos entender se quando o investimento em marketing
# aumenta, as vendas mensais também aumentam.
#
# Pelo modelo de Cleveland-McGill, a comparação pela posição dos pontos no gráfico (no eixo X e Y)
# é bem precisa, então conseguimos perceber facilmente se existe uma tendência de crescimento.
#
# Pela Gestalt, como os pontos ficam espalhados no gráfico, a gente consegue enxergar se eles formam
# um "caminho" ou um padrão, o que indica correlação entre investimento e vendas.
#
# E seguindo o Tufte, o gráfico de dispersão é simples e mostra diretamente a informação principal,
# sem precisar de muitos detalhes extras, deixando claro se existe relação ou não.

# Seu código Python para o gráfico:
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_marketing_vendas, x='Investimento_Marketing', y='Vendas_Mensais')

plt.title("Relação entre Investimento em Marketing e Vendas Mensais")
plt.xlabel("Investimento em Marketing (R$)")
plt.ylabel("Vendas Mensais (R$)")

plt.tight_layout()
plt.show()
```

Desafio 2.1: Vendas Trimestrais com Eixo Y Enganoso

Cenário: Um gerente de vendas recebeu este gráfico e está preocupado com uma suposta queda drástica nas vendas do Q4.

```python
# Dados para o Desafio 2.1
dados_trimestrais = {
    'Trimestre': ['Q1', 'Q2', 'Q3', 'Q4'],
    'Vendas': [1000, 1050, 1020, 980]
}
df_trimestral = pd.DataFrame(dados_trimestrais)

plt.figure(figsize=(7, 5))
ax = sns.barplot(x='Trimestre', y='Vendas', data=df_trimestral, color='lightcoral')
ax.set_ylim(bottom=950) # Eixo Y começando alto
ax.set_title('Vendas Trimestrais - Último Ano', fontsize=14, fontweight='bold')
ax.set_xlabel('Trimestre')
ax.set_ylabel('Vendas (Unidades)')
ax.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# 1. Problemas identificados:
#    a) O eixo Y começa em 950, e isso distorce a visualização. Parece que a diferença entre os trimestres
#       é muito maior do que realmente é. Em gráfico de barras, o ideal é começar o eixo em 0 para não enganar.
#
#    b) O gráfico tem elementos visuais desnecessários, como grid muito forte e título em negrito exagerado,
#       o que pode distrair a atenção do dado principal (as vendas).
#
#    c) A cor "lightcoral" não é um problema grave, mas pode chamar atenção demais sem necessidade. Um tom mais neutro
#       deixaria o gráfico mais profissional e mais fácil de ler.

# 2. Seu código Python refatorado:
plt.figure(figsize=(7, 5))

ax = sns.barplot(x='Trimestre', y='Vendas', data=df_trimestral, color='skyblue')
ax.set_ylim(0, df_trimestral['Vendas'].max() + 100)

ax.set_title('Vendas Trimestrais - Último Ano', fontsize=13)
ax.set_xlabel('Trimestre')
ax.set_ylabel('Vendas (Unidades)')
ax.grid(axis='y', linestyle='--', alpha=0.4)

sns.despine()
plt.tight_layout()
plt.show()
```

Desafio 2.2: Comparação de Produtos com Gráfico de Pizza Lotado

Cenário: A equipe de produto quer ver a proporção de vendas de 8 produtos diferentes, mas o gráfico atual está confuso.

```python
# Dados para o Desafio 2.2
dados_produtos = {
    'Produto': [f'Produto {i}' for i in range(1, 9)],
    'Vendas': [150, 120, 80, 70, 60, 50, 40, 30]
}
df_produtos = pd.DataFrame(dados_produtos)

plt.figure(figsize=(8, 8))
plt.pie(df_produtos['Vendas'], labels=df_produtos['Produto'], autopct='%1.1f%%', startangle=90)
plt.title('Proporção de Vendas por Produto', fontsize=14, fontweight='bold')
plt.show()
```

Desafio 2.4: Gráfico de Linhas com Chartjunk Excessivo

Cenário: Um analista de dados criou este gráfico para mostrar a evolução de duas métricas ao longo do tempo, mas ele está muito poluído.

```python
# Dados para o Desafio 2.4
datas_mes = pd.date_range(start='2023-01-01', periods=12, freq='M')
np.random.seed(42)
metrica_a = np.random.randint(50, 150, 12).cumsum()
metrica_b = np.random.randint(40, 130, 12).cumsum()

df_metricas = pd.DataFrame({
    'Data': datas_mes,
    'Metrica_A': metrica_a,
    'Metrica_B': metrica_b
})

plt.figure(figsize=(10, 6))
plt.plot(df_metricas['Data'], df_metricas['Metrica_A'], label='Métrica A', marker='o', linestyle='-', linewidth=2, color='blue')
plt.plot(df_metricas['Data'], df_metricas['Metrica_B'], label='Métrica B', marker='x', linestyle='--', linewidth=2, color='red')

plt.title('Evolução das Métricas A e B', fontsize=16, fontweight='bold')
plt.xlabel('Data', fontsize=12)
plt.ylabel('Valor', fontsize=12)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True, linestyle=':', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 1. Problemas identificados:
#    a) O gráfico tem muitos detalhes desnecessários, e isso deixa a visualização poluída e mais difícil de interpretar.
#
#    b) A legenda está fora do gráfico, ocupando espaço e deixando o layout estranho. Como só existem
#       duas linhas, a legenda pode ficar dentro do gráfico.
#
#    c) O título está grande e muito chamativo, isso acaba chamando mais atenção
#       do que os próprios dados.

# 2. Seu código Python refatorado:
plt.figure(figsize=(10, 5))

plt.plot(df_metricas['Data'], df_metricas['Metrica_A'], label='Métrica A')
plt.plot(df_metricas['Data'], df_metricas['Metrica_B'], label='Métrica B')

plt.title('Evolução das Métricas A e B', fontsize=13)
plt.xlabel('Data')
plt.ylabel('Valor')

plt.legend()
plt.xticks(rotation=45)

sns.despine()
plt.tight_layout()
plt.show()
```

Desafio 2.5: Comparação de Desempenho com Gráfico de Bolhas Confuso

Cenário: Um gerente de produto quer comparar o desempenho de diferentes funcionalidades (Funcionalidade 1 a 5) com base em duas métricas: 'Uso' e 'Satisfação'. O tamanho da bolha representa o 'Impacto'.

```python
# Dados para o Desafio 2.5
dados_funcionalidades = {
    'Funcionalidade': [f'Func {i}' for i in range(1, 6)],
    'Uso': [0.8, 0.6, 0.9, 0.4, 0.7],
    'Satisfacao': [0.9, 0.7, 0.8, 0.5, 0.6],
    'Impacto': [100, 50, 120, 30, 80]
}
df_funcionalidades = pd.DataFrame(dados_funcionalidades)

plt.figure(figsize=(8, 8))
plt.scatter(x=df_funcionalidades['Uso'], y=df_funcionalidades['Satisfacao'],
            s=df_funcionalidades['Impacto']*5, # Multiplicar para bolhas maiores
            alpha=0.7,
            c=np.random.rand(len(df_funcionalidades), 3)) # Cores aleatórias

for i, row in df_funcionalidades.iterrows():
    plt.text(row['Uso'], row['Satisfacao'], row['Funcionalidade'], fontsize=9, ha='center', va='center')

plt.title('Desempenho das Funcionalidades', fontsize=14, fontweight='bold')
plt.xlabel('Uso (0-1)', fontsize=12)
plt.ylabel('Satisfação (0-1)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)
plt.show()
```

```python
# 1. Problemas identificados:
#    a) O gráfico usa cores aleatórias, mas essas cores não significam nada. Isso pode confundir.
#
#    b) O tamanho das bolhas está muito exagerado (['Impacto']*5), e isso pode deixar o gráfico difícil de ler,
#       porque algumas bolhas ficam grandes demais e escondem outras.
#
#    c) Os textos com o nome das funcionalidades ficam em cima das bolhas e podem se sobrepor.

# 2. Seu código Python refatorado:
plt.figure(figsize=(8, 6))

plt.scatter(
    x=df_funcionalidades['Uso'],
    y=df_funcionalidades['Satisfacao'],
    s=df_funcionalidades['Impacto'] * 2,  # tamanho mais controlado
    alpha=0.7,
    color='skyblue'
)

# Colocar os nomes um pouco acima do ponto para não ficar em cima da bolha
for i, row in df_funcionalidades.iterrows():
    plt.text(row['Uso'], row['Satisfacao'] + 0.02, row['Funcionalidade'], fontsize=9, ha='center')

plt.title('Desempenho das Funcionalidades', fontsize=13)
plt.xlabel('Uso (0-1)')
plt.ylabel('Satisfação (0-1)')

plt.grid(True, linestyle='--', alpha=0.3)
sns.despine()
plt.tight_layout()
plt.show()
```
